```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(1,'/Users/user/Documents/MeerFish')
import cosmo
import model
import fisher
```

# Replicating Alex Barriera's Fig 1 plot in https://arxiv.org/pdf/2302.09066.pdf

```python
### params used in Alex's paper:
###  - see end of Sec 1
z = 1
Vlen = 100**(1/3) * 1e3 # box side length in Mpc/h for 100 (Gpc/h)^3 box
V_bin = Vlen**3 # equivalent to 100Gpc * 1e9
theta_FWHM = 0
sigma_z = 0
Tbar = 1

### k-bins:
kmin = np.pi/V_bin**(1/3)
kmax = 0.2
#kbins = np.logspace(np.log10(kmin),np.log10(kmax),41) 
k = np.logspace(np.log10(kmin),np.log10(kmax),40) # log bins not currently working in Fisher ell code (use below if ell)
#kbins = np.linspace(kmin,kmax,100)
#k = (kbins[1:] + kbins[:-1])/2 #centre of k bins

### Cosmological parameters:
Tbar1,Tbar2,b1,b2,bphi1,bphi2,f,a_perp,a_para,A_BAO,f_NL = cosmo.SetCosmology(z=z,return_cosmopars=True) # set initial default cosmology
f_NL = 5
cosmopars = np.array([Tbar1,Tbar2,b1,b2,bphi1,bphi2,f,a_perp,a_para,A_BAO,f_NL])
Pmod = cosmo.MatterPk(z)
b_1 = 2
b_phi = 3.37 # equiv to cosmo.b_phi_universality(b_1)
print(cosmo.b_phi_universality(b_1))
```

```python
theta_ids = [r'$f_{\rm NL}$']
#ells = [0]

nbars = [2, 2e-4] # [optimistic case, realistic case]
theta_angs = np.linspace(-np.pi/2,np.pi/2,60)
for n in range(len(nbars)):
    theta = np.radians(73.5) # theta = 73.5 gives ~universality line in Alex's plot
    b_1A = b_1 + np.cos(theta + np.pi)
    b_1B = b_1 + np.cos(theta)
    b_phiA = b_phi + np.sin(theta + np.pi)
    b_phiB = b_phi + np.sin(theta)
    plt.plot([b_1A,b_1B],[b_phiA,b_phiB],ls='--',color='gray')

    b_1A,b_1B,b_phiA,b_phiB,ratio = [],[],[],[],[]
    for i in range(len(theta_angs)):
        b_1A.append(b_1 + np.cos(theta_angs[i] + np.pi))
        b_1B.append(b_1 + np.cos(theta_angs[i]))
        b_phiA.append(b_phi + np.sin(theta_angs[i] + np.pi))
        b_phiB.append(b_phi + np.sin(theta_angs[i]))
        
        cosmopars = np.array([Tbar,Tbar,b_1,b_1,b_phi,b_phi,f,a_perp,a_para,A_BAO,f_NL])
        surveypars = z,V_bin,V_bin,V_bin,theta_FWHM,theta_FWHM,sigma_z,sigma_z,1/nbars[n],1/nbars[n]
        #F_ST = fisher.Matrix_ell(theta_ids,k,Pmod,cosmopars,surveypars,ells,tracer='1')
        F_ST = fisher.Matrix_2D(theta_ids,k,Pmod,cosmopars,surveypars,tracer='1')

        cosmopars = np.array([Tbar,Tbar,b_1A[i],b_1B[i],b_phiA[i],b_phiB[i],f,a_perp,a_para,A_BAO,f_NL])
        surveypars = z,V_bin,V_bin,V_bin,theta_FWHM,theta_FWHM,sigma_z,sigma_z,1/(nbars[n]/2),1/(nbars[n]/2)
        #F_MT = fisher.Matrix_ell(theta_ids,k,Pmod,cosmopars,surveypars,ells,tracer='MT')
        F_MT = fisher.Matrix_2D(theta_ids,k,Pmod,cosmopars,surveypars,tracer='MT')
        ratio.append( (1/np.sqrt(F_MT[-1,-1])) / (1/np.sqrt(F_ST[-1,-1])) )

    if n==0: ratio = np.log10(ratio)
    plt.scatter(b_1A,b_phiA,c=ratio,s=100,cmap='jet')
    plt.scatter(b_1B,b_phiB,c=ratio,s=100,cmap='jet')
    if n==0: plt.colorbar(label=r'${\rm log}_{10} \, \sigma_{f_{\rm NL}}^{\rm MT} / \sigma_{f_{\rm NL}}^{\rm ST}$')
    if n==1: plt.colorbar(label=r'$\sigma_{f_{\rm NL}}^{\rm MT} / \sigma_{f_{\rm NL}}^{\rm ST}$')
    plt.xlabel(r'$b_1$')
    plt.ylabel(r'$b_\phi$')
    plt.show()
```
